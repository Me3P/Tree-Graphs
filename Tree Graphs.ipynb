```python
import networkx as nx
import itertools

def bfs_trees(G):
    n=G.order()
    adj = G.adjacency_matrix()
    G1 = G.networkx_graph() 
    BFS_trees=[] #This holds all BFS TREES AT THE END
    all_permutations=list(itertools.permutations([i for i in range(1,n)]))
    for permutation in all_permutations:
        mapping=[0]*n
        for i in range (1,n):
            mapping[i]=permutation[i-1]
        G1=nx.Graph()
        for i in range(n):
            G1.add_node(mapping[i])
        for i in range(n):
            for j in range(n):
                if adj[mapping[i]][mapping[j]]==1:
                    G1.add_edge(mapping[i],mapping[j])
        t1=nx.bfs_tree(G1,0)
        BFS_trees.append(t1)
    A=[tuple(Graph(nx.to_dict_of_dicts(x.to_undirected())).edges(labels=False)) for x in BFS_trees]
    return list(set(A))
```
